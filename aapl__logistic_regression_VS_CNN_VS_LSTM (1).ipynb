Three AI techniques are selected for evaluation in this analysis:  
1. **Logistic Regression** (Regression-based technique)  
2. **Convolutional Neural Network (CNN)** (Deep Learning approach for short-term trend detection)  
3. **Long Short-Term Memory Network (LSTM)** (Deep Learning approach for sequential long-term trend modeling).  

---

### **1. Logistic Regression**  

**Description**:  
Logistic Regression is a supervised machine learning technique used for binary classification problems. In stock price forecasting, it can predict the **direction of price movement** (up or down) based on lagged returns as input features.

**Input Data**:  
- Features: Lagged returns as deminstrated in the code section of logistic regression.  
- Target: Binary direction (`1` for upward movement, `0` for downward/no movement).  

**Expected Output**:  
- A binary classification output predicting whether the stock price will rise or fall.  

**Strengths**:  
- **Simplicity**: Easy to implement and interpret.  
- **Speed**: Computationally efficient, requiring less training time.  
- **Feature Interpretability**: Coefficients provide insights into the importance of features.  

**Weaknesses**:  
- **Linear Assumption**: Assumes a linear relationship between inputs and outputs, which may not hold in financial time series.  
- **Limited Complexity**: Struggles to model non-linear patterns or relationships in stock price data.  
- **Imbalanced Predictions**: Tends to favor upward price movements due to their dominance in historical data.  

**Application Example**:  
- Predicting whether Apple stock's closing price will increase or not based on lagged returns from the previous 5 days.  
- **Performance**: Achieved ~51.5% accuracy, slightly above random chance.  

---

### **2. Convolutional Neural Network (CNN)**  

**Description**:  
CNNs are a class of deep learning models typically used in computer vision, but they can effectively extract **short-term patterns** in sequential time series data like stock prices. They identify local dependencies through convolutional filters.

**Input Data**:  
- Features: A sliding window of stock closing prices (e.g., 60 consecutive days).  
- Target: The next day’s closing price.  

**Expected Output**:  
- A continuous numerical prediction of the next day’s stock closing price.  

**Strengths**:  
- **Short-Term Pattern Detection**: Captures local trends and short-term dependencies using convolutional filters.  
- **Efficiency**: Faster to train compared to RNN-based models like LSTMs.  
- **Good Baseline Model**: Performs well when stock price trends are dominated by short-term movements.  

**Weaknesses**:  
- **Limited Long-Term Memory**: Cannot model long-term dependencies effectively.  
- **Data Sensitivity**: Performance heavily depends on the choice of sliding window size and filter configurations.  

**Application Example**:  
- Using a 60-day window of Apple stock closing prices to predict the next day’s closing price.  
- CNN captured short-term price movements efficiently and provided reasonable predictions.  

---

### **3. Long Short-Term Memory Network (LSTM)**  

**Description**:  
LSTM is a type of Recurrent Neural Network (RNN) designed to capture **long-term dependencies** in sequential data. It overcomes the limitations of traditional RNNs by using special gates (input, forget, and output gates) to regulate information flow over time.

**Input Data**:  
- Features: A sliding window of stock closing prices (previous 60 days).  
- Target: The next day’s closing price.  

**Expected Output**:  
- A continuous numerical prediction of the next day’s stock closing price.  

**Strengths**:  
- **Long-Term Pattern Detection**: Effectively models complex temporal relationships in financial time series.  
- **Handles Non-Linearity**: Captures non-linear patterns and trends in stock prices.  
- **Robust for Volatile Data**: Suitable for highly dynamic and noisy financial markets.  

**Weaknesses**:  
- **High Computational Cost**: Slower to train compared to CNN due to sequential dependencies.  
- **Hyperparameter Sensitivity**: Requires careful tuning of parameters (including number of units and learning rate) to achieve optimal performance.  

**Application Example**:  
- Using 60 days of Apple stock closing prices to predict the next day’s closing price.  
- LSTM achieved strong performance by modeling long-term trends and temporal patterns.

```python
# Import Libraries
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
```



# 1. Data Preprocessing

The data preprocessing in this case includes feature engineering by calculating the logarithmic returns (returns) of the "Close" prices. Log returns stabilize variance and are widely used in financial modeling. A binary direction indicator (direction) is then created, where 1 represents an upward price movement, and -1 represents a downward or no movement.

To incorporate historical trends, the lagit function generates lagged features for the returns and their directions over the past 5 days. These lagged features enable the model to capture short-term patterns, such as momentum or reversals. Rows with missing values caused by lagging are dropped to ensure a clean dataset.

This step transforms the data into a structured format suitable for binary classification.


```python
# Feature Engineering
df['returns'] = np.log(df['Close'].pct_change() + 1)  # Log Returns
df['direction'] = [1 if i > 0 else -1 for i in df['returns']]  # Direction: Up (1), Down (-1)

# Lagged Features
def lagit(df, lags):
    for i in range(1, lags + 1):
        df[f'Lag_{i}'] = df['returns'].shift(i)
        df[f'Lag_{i}_dir'] = [1 if j > 0 else -1 for j in df[f'Lag_{i}']]
    return [f'Lag_{i}_dir' for i in range(1, lags + 1)]

# Create lagged returns
dirnames = lagit(df, 5)

# Drop NaNs caused by lagged features
df.dropna(inplace=True)
```

# 2. Train-Test Split

We split into training and testing sets, preserving the time series order by setting shuffle=False. The lagged features (dirnames) are used as predictors (X), and the binary direction is the target (y). The training set comprises 75% of the data, while the remaining 25% is reserved for testing.

```python
# Define features and target
X = df[dirnames]
y = df['direction']

# Split the data (no shuffle for time series)
train, test = train_test_split(df, shuffle=False, test_size=0.25, random_state=42)

X_train, y_train = train[dirnames], train['direction']
X_test, y_test = test[dirnames], test['direction']
```

# 3. Feature Scaling

Feature scaling is applied using StandardScaler to standardize the predictors (X_train and X_test) to have a mean of 0 and a standard deviation of 1.

```python
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# 4. Hyperparameter Tuning

Hyperparameter tuning is implemented for the logistic regression model using GridSearchCV for optimization purposes. A grid of parameters (C, penalty, and solver) is tested with 5-fold cross-validation on the training data. The model is evaluated using accuracy as the scoring metric.

```python
# Set up GridSearchCV
param_grid = {'C': [0.01, 0.1, 1, 10], 'penalty': ['l1', 'l2'], 'solver': ['liblinear']}
grid_search = GridSearchCV(LogisticRegression(max_iter=500), param_grid, scoring='accuracy', cv=5)
grid_search.fit(X_train_scaled, y_train)

# Best model
best_model = grid_search.best_estimator_
print("Best Parameters:", grid_search.best_params_)
```

The **best parameters** identified through GridSearchCV are:  
- **C = 1**: Regularization strength (inverse of regularization).  
- **penalty = 'l1'**: L1 regularization, which promotes sparsity in coefficients.  
- **solver = 'liblinear'**: A solver compatible with L1 regularization.

These parameters are used to fit the optimized Logistic Regression model.

# 5. Train and Predict

```python
# Fit the best model
best_model.fit(X_train_scaled, y_train)

# Predictions
test['prediction_logit'] = best_model.predict(X_test_scaled)
test['strat_logit'] = test['prediction_logit'] * test['returns']
```

# 6. Model Evaluation

```python
# Confusion Matrix and Classification Report
print("Confusion Matrix:")
print(metrics.confusion_matrix(y_test, test['prediction_logit']))

print("\nClassification Report:")
print(metrics.classification_report(y_test, test['prediction_logit']))

# Accuracy Score
accuracy = metrics.accuracy_score(y_test, test['prediction_logit'])
print(f"Logistic Regression Accuracy: {accuracy:.4f}")
```

The model evaluation results indicate the following:

Overall accuracy: 51.54%.
The model favors predicting upward movements (1) over downward movements (-1), resulting in imbalanced performance.

# CNN and LSTM

The limited performance of Logistic Regression as seen in the previous section highlights its inability to capture complex patterns in stock price movements. We therefore consider CNN and LSTM. CNN models are considered for their ability to extract short-term trends from sequential data, while LSTM models are well-suited for capturing long-term dependencies in time series data. Both models address the limitations of Logistic Regression by leveraging deep learning to better model the non-linear and temporal nature of stock prices.

```python
# Import Libraries
import yfinance as yf
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, Conv1D, Flatten, LSTM
import matplotlib.pyplot as plt
```

# 1. Fetch Data and Preprocess

We preprocess the data by ensuring that MultiIndex columns are flattened by removing the "Ticker" level, keeping only the "Price" level, and removing the column axis name. This simplifies the DataFrame structure for easier access and manipulation of columns.

The conv2value function extracts the "Close" prices as the primary feature for modeling and converts them into a NumPy array.

The scalevalue function applies MinMax scaling to normalize the "Close" prices between 0 and 1.

The data is then split into training and testing sets, with 90% used for training.

Sliding windows of 60 days are created for both sets to form input-output pairs, where the previous 60 days’ prices (x_train, x_test) predict the next day’s price (y_train, y_test). The data is reshaped into a 3D format (samples, timesteps, features) required for CNN and LSTM models.


```python
df_AP = yf.download("AAPL", start="2010-01-01", end="2023-12-31")
```

```python
# Flatten the MultiIndex by keeping only the 'Price' level
df_AP.columns = df_AP.columns.droplevel(level=1)  # Drop the 'Ticker' level
df_AP = df_AP.rename_axis(None, axis=1)  # Remove the column axis name
print(df_AP.head())
```

```python
# Extract Close Prices
def conv2value(df):
    data = df.filter(['Close'])
    dataset = data.values
    return dataset
```

```python
#Scale the Data
def scalevalue(dataset):
    scaler = MinMaxScaler(feature_range=(0, 1))
    scaled_data = scaler.fit_transform(dataset)
    return scaled_data, scaler
```

```python
# Training and Testing Data
def creatdataset(df):
    dataset = conv2value(df)
    scaled_data, scaler = scalevalue(dataset)

    training_data_len = int(np.ceil(len(dataset) * 0.9))  # 90% for training
    train_data = scaled_data[0:training_data_len, :]

    # Create x_train and y_train
    x_train, y_train = [], []
    for i in range(60, len(train_data)):
        x_train.append(train_data[i-60:i, 0])
        y_train.append(train_data[i, 0])

    x_train, y_train = np.array(x_train), np.array(y_train)
    x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1], 1))

    # Test Data
    test_data = scaled_data[training_data_len - 60:, :]
    x_test, y_test = [], scaled_data[training_data_len:, :]
    for i in range(60, len(test_data)):
        x_test.append(test_data[i-60:i, 0])

    x_test = np.array(x_test)
    x_test = np.reshape(x_test, (x_test.shape[0], x_test.shape[1], 1))

    return x_train, y_train, x_test, y_test, scaler

# Create dataset
X_train, y_train, X_test, y_test, scaler = creatdataset(df_AP)
```

# 2. CNN Model

This Convolutional Neural Network (CNN) model in this section is designed for a one-dimensional input sequence with 60 time steps and a single feature. The model consists of an initial 1D convolutional layer with 5 filters, a kernel size of 3, and 'ReLU' activation, followed by a Flatten layer to convert the 2D output to 1D. It then includes a dense layer with 40 neurons using a 'sigmoid' activation function, and a final dense layer with a single neuron. The model is compiled with the Adam optimizer and Mean Squared Error (MSE) loss function. Training is performed over 20 epochs, with performance validated on a separate test dataset, and predictions are subsequently transformed back to their original scale for evaluation.

```python
print("Training CNN Model...")
cnn_model = Sequential()
cnn_model.add(Conv1D(5, 3, strides=1, activation='relu', input_shape=(60, 1)))
cnn_model.add(Flatten())
cnn_model.add(Dense(40, activation='sigmoid'))
cnn_model.add(Dense(1, activation='sigmoid'))
cnn_model.compile(optimizer='adam', loss='mse')
cnn_model.summary()

# Train CNN Model
history_cnn = cnn_model.fit(X_train, y_train, epochs=20, validation_data=(X_test, y_test), shuffle=False, verbose=2)

# Predictions
y_pred_cnn = cnn_model.predict(X_test)
y_actual = pd.Series(scaler.inverse_transform(y_test.reshape(-1, 1)).reshape(-1))
y_pred_cnn = pd.Series(scaler.inverse_transform(y_pred_cnn.reshape(-1, 1)).reshape(-1))
```

```python
# Plot Training History
plt.figure(figsize=(14, 7))
plt.plot(history_cnn.history['loss'], label='Train Loss')
plt.plot(history_cnn.history['val_loss'], label='Validation Loss')
plt.title('CNN Model Training Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

Throughout the 20 epochs of training the CNN model, there was a reduction in both the training loss and the validation loss initially. The validation loss also followed a similar pattern. However, after the initial epochs, the improvement in the validation loss plateaued, suggesting that the model reached a point where further training did not lead to significant validation performance gains.

```python
# Visualize Results
plt.figure(figsize=(16, 6))
plt.plot(y_actual, label='Actual Prices', color='blue')
plt.plot(y_pred_cnn, label='CNN Predictions', color='orange')
plt.title('CNN Model - Actual vs Predicted Prices')
plt.xlabel('Days')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

We can see that the CNN model's predictions closely follow the actual prices. However, slight deviations between the lines suggest areas where the model's predictions could be further refined for improved accuracy.

# 3. LSTM Model

We implement the LSTM model using Keras to predict AAPL stock price. We develop the model using a sequential architecture with the following layers:

LSTM Layer (128 units): The first LSTM layer with 128 units and return_sequences=True to output the full sequence for further LSTM layers.

LSTM Layer (64 units): The second LSTM layer with 64 units and return_sequences=False to output the final hidden state.

Dense Layer (25 units): A fully connected layer with 25 neurons.

Dense Layer (1 unit): The final output layer with a single neuron to predict the target variable.

The model is compiled using the Adam optimizer and Mean Squared Error (MSE) loss function.

```python
print("Training LSTM Model...")
lstm_model = Sequential()
lstm_model.add(LSTM(128, return_sequences=True, input_shape=(X_train.shape[1], 1)))
lstm_model.add(LSTM(64, return_sequences=False))
lstm_model.add(Dense(25))
lstm_model.add(Dense(1))
lstm_model.compile(optimizer='adam', loss='mean_squared_error')
lstm_model.summary()

# Train LSTM Model
history_lstm = lstm_model.fit(X_train, y_train, batch_size=32, epochs=20, validation_data=(X_test, y_test))

# Predictions
y_pred_lstm = lstm_model.predict(X_test)
y_pred_lstm = pd.Series(scaler.inverse_transform(y_pred_lstm.reshape(-1, 1)).reshape(-1))
```

Throughout the 20 epochs the LSTM model shows a significant reduction in training  and validation loss. These results indicate effective learning and strong generalization, with the model accurately predicting on unseen data throughout the training process. The occasional fluctuations in validation loss reflect minor adjustments for improved overall performance.

```python
# Visualize Results
plt.figure(figsize=(16, 6))
plt.plot(y_actual, label='Actual Prices', color='blue')
plt.plot(y_pred_lstm, label='LSTM Predictions', color='green')
plt.title('LSTM Model - Actual vs Predicted Prices')
plt.xlabel('Days')
plt.ylabel('Close Price')
plt.legend()
plt.show()
```

 The graph indciates close alignment between the model predictions and the actual prices. The close overlap, especially in the latter part of the graph, suggests that the model's accuracy improves over time, making it a reliable tool for predicting stock prices in this period.
